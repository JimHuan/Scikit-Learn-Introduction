####  (seletion sorting)
    
    
        n
        1221n
        2332n
        n-1  n
     O(n^2)

```python
import numpy as np 
sorting_list = np.random.normal(loc = 3, size = 10)
print(sorting_list)  #
```

```python
#1i
def seletion_sorting1(array):
    n = len(array)
    for i in range(n-1):
        for j in range(i+1, n):
            if array[i] > array[j]:
                array[i], array[j] = array[j], array[i]
    return array
```

```python
#2:ii
def seletion_sorting2(array):
    n = len(array)
    for i in range(n-1):
        min = i
        for j in range(i, n):
            if array[min] > array[j]:
                min  = j
        array[i], array[min]  = array[min] , array[i]      
    return array
```

```python
#
%timeit seletion_sorting1(sorting_list)
```

```python
%timeit seletion_sorting2(sorting_list)
```

#### (buble sorting
    

```python
def buble_sorting(array):
    n = len(sorting_list)
    for i in range(n-1,0, -1): #n-1
        for j in range(0, i):
            if array[j] > array[j+1]:
                array[j], array[j+1] = array[j+1], array[j]
    return array
```

```python
%timeit buble_sorting(sorting_list)
```

#### (insertion sorting)
    
    O(n^2)

```python
def insertion_sorting(array):
    n = len(array)
    for i in range(1, n):
        for j in range(i, 0, -1):
            if array[j] < array[j -1]:
                array[j], aray[j-1] = array[j-1], array[j]
            else:
                break
    return array
```

```python
%timeit insertion_sorting(sorting_list)
insertion_sorting(sorting_list)
```

#### (quick sorting)
    
    O(nlogn)

```python
def quick_sorting(array):
    n = len(array)
    if n <= 1:
        return array
    smaller = []
    bigger = []
    bench = array[0]
    for i in range(1, n):
        if array[i] < bench:
            smaller.append(array[i])
        else:
            bigger.append(array[i])
    return quick_sorting(smaller) + [bench] + quick_sorting(bigger)  
```

```python
%timeit quick_sorting(sorting_list)
```
