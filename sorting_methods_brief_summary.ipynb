### python实现排序算法
    时间复杂度只有在大量数据条件下才能表现出来，在少量数据下比较时间复杂度没意义。


#### 一、 选择排序(seletion sorting)
    顾名思义，直接选择大（小）的数排序。
    从小到大排序为例具体实现：
        假设某个序列具有n个数据；
        ①第1位跟第2位比较，若第2位小，则互换位置，否则不操作；重复第1位与后面数据比较至第n位；
        ②第2位跟第3位比较，若第3位小，则互换位置，否则不操作；重复第2位与后面数据比较至第n位；
        按照①②的步骤直至第n-1位 与 第n位的比较
     时间复杂度O(n^2)

```python
import numpy as np 
sorting_list = np.random.normal(loc = 3, size = 10)
print(sorting_list)  #产生正态分布的随机序列
```

```python
#写法1：把第i位在与后面序列比较中及时更新，时刻保持最小
def seletion_sorting1(array):
    n = len(array)
    for i in range(n-1):
        for j in range(i+1, n):
            if array[i] > array[j]:
                array[i], array[j] = array[j], array[i]
    return array
```

```python
#写法2:在第i位到最后一位中，选出最小的值，再与第i位互换
def seletion_sorting2(array):
    n = len(array)
    for i in range(n-1):
        min = i
        for j in range(i, n):
            if array[min] > array[j]:
                min  = j
        array[i], array[min]  = array[min] , array[i]      
    return array
```

```python
#测试
%timeit seletion_sorting1(sorting_list)
```

```python
%timeit seletion_sorting2(sorting_list)
```

#### 二、冒泡算法(buble sorting）
    逐位两两比较，把大数往右移。

```python
def buble_sorting(array):
    n = len(sorting_list)
    for i in range(n-1,0, -1): #n-1是为了保证序列最后一位被索引
        for j in range(0, i):
            if array[j] > array[j+1]:
                array[j], array[j+1] = array[j+1], array[j]
    return array
```

```python
%timeit buble_sorting(sorting_list)
```

#### 三、快速插入排序(insertion sorting)
    将待排序的数逐一与前面的数进行比较排序，插入合适的位置；一直操作至最后一个数。
    时间复杂度为O(n^2)

```python
def insertion_sorting(array):
    n = len(array)
    for i in range(1, n):
        for j in range(i, 0, -1):
            if array[j] < array[j -1]:
                array[j], aray[j-1] = array[j-1], array[j]
            else:
                break
    return array
```

```python
%timeit insertion_sorting(sorting_list)
insertion_sorting(sorting_list)
```

#### 四、快速排序(quick sorting)
    随机（第一个或者最后一个）选择一个数作为基准，将其他数分成比基准大和比基准小的两组，重新排序；重复操作两组新序列，直至新的序列只剩一个数。
    时间复杂度O(nlogn)

```python
def quick_sorting(array):
    n = len(array)
    if n <= 1:
        return array
    smaller = []
    bigger = []
    bench = array[0]
    for i in range(1, n):
        if array[i] < bench:
            smaller.append(array[i])
        else:
            bigger.append(array[i])
    return quick_sorting(smaller) + [bench] + quick_sorting(bigger)  
```

```python
%timeit quick_sorting(sorting_list)
```
